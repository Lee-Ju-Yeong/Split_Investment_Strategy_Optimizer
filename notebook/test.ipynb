```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load the CSV data
file_path = 'E:/AI/pythonProject/venv/masicsplit/parameter_simulation/parameter_combinations_20240801_2341.csv'
data = pd.read_csv(file_path)

# Extract relevant columns for correlation analysis
relevant_columns = [
    'num_splits', 'buy_threshold', 'investment_ratio', 'max_stocks',
    'per_threshold', 'pbr_threshold', 'dividend_threshold', 'min_additional_buy_drop_rate',
    'Average_Total_Value', 'Average_CAGR', 'Average_MDD', 'sharp'
]

# Standardization
scaler = StandardScaler()
scaled_data_standard = scaler.fit_transform(data[relevant_columns])

# Min-Max Scaling
minmax_scaler = MinMaxScaler()
scaled_data_minmax = minmax_scaler.fit_transform(data[relevant_columns])

# Convert scaled data back to DataFrame for easier handling
scaled_df_standard = pd.DataFrame(scaled_data_standard, columns=relevant_columns)
scaled_df_minmax = pd.DataFrame(scaled_data_minmax, columns=relevant_columns)

# Compute the correlation matrix for Standard Scaled data
correlation_matrix_standard = scaled_df_standard.corr()

# Compute the correlation matrix for Min-Max Scaled data
correlation_matrix_minmax = scaled_df_minmax.corr()

# Plot the heatmap for Standard Scaled data
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix_standard, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Parameters (Standard Scaled)')
plt.show()

# Plot the heatmap for Min-Max Scaled data
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix_minmax, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix of Parameters (Min-Max Scaled)')
plt.show()
```
